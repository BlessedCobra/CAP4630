### Imports

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

# **Function 1: Local Minimum**

### Function: f(x,y) = 4x^2 + 4y^2

```
def f1(x1, y1):
    return (4*x1*x1) + (4*y1*y1)
```

```
x1 = np.linspace(-2, 2, 30)
y1 = np.linspace(-2, 2, 30)

X1, Y1 = np.meshgrid(x1, y1)
Z1 = f1(X1, Y1)
```

### Plot the region containing the local minimum, which is f(0,0)

```
plt.imshow(Z1, extent=[-2, 2, -2, 2], origin='lower',
           cmap='RdBu')
plt.colorbar()
plt.axis(aspect='image');
```

# **Function 2: Local Maximum**

### Function: f(x,y) = -(4x^2 + 4y^2)

```
def f2(x2, y2):
    return -((4*x2*x2) + (4*y2*y2))
```

```
x2 = np.linspace(-2, 2, 30)
y2 = np.linspace(-2, 2, 30)

X2, Y2 = np.meshgrid(x2, y2)
Z2 = f2(X2, Y2)
```

### Plot the region containing the local maximum, which is f(0,0)

```
plt.imshow(Z2, extent=[-2, 2, -2, 2], origin='lower',
           cmap='RdBu')
plt.colorbar()
plt.axis(aspect='image');
```

# **Function 3: Saddle Point**


### Function: f(x, y) = 4x^2 - 4y^2

```
def f3(x3, y3):
    return (4*x3*x3) - (4*y3*y3)
```

```
x3 = np.linspace(-2, 2, 300)
y3 = np.linspace(-2, 2, 300)

X3, Y3 = np.meshgrid(x3, y3)
Z3 = f3(X3, Y3)
```

### Plot the region containing the saddle point, which is f(0,0)

```
plt.imshow(Z3, extent=[-2, 2, -2, 2], origin='lower',
           cmap='seismic')
plt.colorbar()
plt.axis(aspect='image');
```

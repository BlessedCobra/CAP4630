### Imports

```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

### Initialize x1, x2, and y parameters, and plot them

```
x1s = 2 * np.random.rand(100, 1)
x2s = 2 * np.random.rand(100, 1)
ys = 4 + (3 * x1s + np.random.rand(100, 1)) + (3 * x2s + np.random.rand(100, 1))
```

```
ax = plt.axes(projection='3d')
ax.scatter3D(x1s, x2s, ys, cmap='Greens')
```

```
# training set
train_x1s = x1s[:80]
train_x2s = x2s[:80]
train_ys = ys[:80]
# test set
test_x1s = x1s[80:]
test_x2s = x2s[80:]
test_ys = ys[80:]
```

# Linear Regression Time

```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# initial value for weight w1 and w2, and bias b
w1 = np.random.randn(1)
w2 = np.random.rand(1)
b = np.zeros(1)
```

```
# gradient descent
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = (w1 * train_x1s[i]) + (w2 * train_x2s[i]) + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_x1s[i]
    grad_w2 = (y_pred - train_ys[i]) * train_x2s[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
# compute the test_loss 
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (((( w1 * test_x1s[i] ) + ( w2 * test_x2s[i] ) + b) - test_ys[i] )**2)
test_loss /= 20
```

```
test_loss
```

```
# our predicted values for y
pred_ys = (w1 * test_x1s) + (w2 * test_x2s) + b
```

### Plot test values (blue) and predicted values (orange)

```
ax = plt.axes(projection='3d')
ax.scatter3D(test_x1s, test_x2s, test_ys)
ax.scatter3D(test_x1s, test_x2s, pred_ys)
```
